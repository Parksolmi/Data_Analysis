<h3> 1. 실버타운 데이터 전처리 </h3>

- 불필요한 칼럼 지우기 : 'field3', 'field4', 'field5', '_GC_TYPE'
- 칼럼명 바꾸기
- 시설 유형이 '실버타운'에 해당하지 않는 데이터 지우기


```python
silver_town_df = silver_town_df.drop(columns=['field3', 'field4', 'field5', '_GC_TYPE'])
```

```python
silver_town_df = silver_town_df.rename(columns={
    'field1': '시설명',
    'field2': '시설유형',
    '_CLEANADDR': '주소',
    '_X': '경도',
    '_Y': '위도'
})
silver_town_df.head()
```

```python
unique_types = silver_town_df['시설유형'].unique()
print(unique_types)
```

```python
silver_town_df = silver_town_df[silver_town_df['시설유형'] != '스파게티,파스타전문']
```

```python
unique_types = silver_town_df['시설유형'].unique()
print(unique_types)
```

<h3> 2. 국내생활편의시설 데이터 전처리 </h3>

- 불필요한 칼럼 지우기
  - LCLAS_NM(대분류명), MLSFC_NM(중분류명), POI_ID(POI_ID), POI_NM(POI명), ASSTN_NM(보조명), LI_NM(리명), LNBR_NO(번지번호), LEGALDONG_CD(법정동코드), ADSTRD_CD(행정동코드), RDNMADR_CD(도로명주소), GID_CD(격자코드), LAST_CHG_DE(최종변경일자), ORIGIN_NM(출처명), FILE_NM(파일명), BASE_DE(기준일자)
- 주소 데이터 하나의 칼럼으로 합치기
  - CTPRVN_NM(시도명), SIGNGU_NM(시군구명), LEGALDONG_NM(법정동명), RDNMADR_NM(도로명주소명), BULD_NO(건물번호)
- 칼럼명 바꾸기

```python
conveninent_facilities_df = conveninent_facilities_df.drop(columns=['LCLAS_NM', 'MLSFC_NM', 'POI_ID', 'POI_NM', 'ASSTN_NM', 'LI_NM', 'LNBR_NO', 'LEGALDONG_CD', 'ADSTRD_CD', 'RDNMADR_CD', 'GID_CD', 'LAST_CHG_DE', 'ORIGIN_NM', 'FILE_NM', 'BASE_DE'])
conveninent_facilities_df.head()
```

```python
# 여러 칼럼을 하나의 'Address' 칼럼으로 합치기
conveninent_facilities_df['Address'] = conveninent_facilities_df.apply(lambda row: f"{row['CTPRVN_NM']} {row['SIGNGU_NM']} {row['LEGALDONG_NM']} {row['RDNMADR_NM']} {row['BULD_NO']}", axis=1)

# 원래 칼럼 삭제
conveninent_facilities_df = conveninent_facilities_df.drop(columns=['CTPRVN_NM', 'SIGNGU_NM', 'LEGALDONG_NM', 'RDNMADR_NM', 'BULD_NO'])

conveninent_facilities_df.head()
```

```python
conveninent_facilities_df = conveninent_facilities_df.rename(columns={
    'ID': '분류아이디',
    'BHF_NM': '지점명',
    'CL_CD': '분류코드',
    'CL_NM': '분류명',
    'PNU': '지번고유코드',
    'LC_LO': '경도',
    'LC_LA': '위도',
    'Address': '주소',
})
conveninent_facilities_df.head()
```

<h3> 3. 전국 병의원 및 약국 현황 데이터 데이터 전처리 </h3>

- 불필요한 칼럼 지우기
  - 암호화요양기호, 종별코드, 시도코드, 시도코드명, 시군구코드, 시군구코드명, 읍면동, 전화번호, 병원홈페이지, 개설일자, 의과일반의 인원수, 의과인턴 인원수, 의과레지턴트 인원수, 의과전문의 인원수, 치과일반의 인원수, 치과인턴인원수, 치과레지턴트 인원수, 치과전문의 인원수, 한방일반의 인원수, 한방인턴 인원수, 한방레지턴트 인원수, 한방전문의 인원수, 조산사 인원수
- 칼럼명 바꾸기

```python
# 칼럼 삭제
medical_facilities_df = medical_facilities_df.drop(columns=['암호화요양기호', '종별코드', '시도코드', '시도코드명', '시군구코드', '시군구코드명', '읍면동', '전화번호', '병원홈페이지', '개설일자', '의과일반의 인원수', '의과인턴 인원수', '의과레지던트 인원수', '의과전문의 인원수', '치과일반의 인원수', '치과인턴 인원수', '치과레지던트 인원수', '치과전문의 인원수', '한방일반의 인원수', '한방인턴 인원수', '한방레지던트 인원수', '한방전문의 인원수', '조산사 인원수'])
medical_facilities_df.head()
```

```python
medical_facilities_df = medical_facilities_df.rename(columns={
    '좌표(X)': '경도',
    '좌표(Y)': '위도',
})
medical_facilities_df.head()
```
