```python
mapCB=folium.Map(location=[37.491145,126.9737511], zoom_start=15)
```

```python
for i, store in mediGeoData.iterrows():
  folium.Marker(location=[store['_Y'],store['_X']],

popup=store['field2'],icon=folium.Icon(color='red',icon='star')).add_to(mapCB)
mapCB.save('/content/drive/MyDrive/Colab Notebooks/Github/Data_Analysis/MediGeoData/map.html')
```
